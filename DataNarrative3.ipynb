```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<h3>AusOpen-men-2013</h3>

```python
df = pd.read_csv("AusOpen-men-2013.csv")
df.head()
df.columns
```

Question  1: Is there a relationship between a player's unforced errors commited by the player 1 (UFE.1) and their success in the tournament, and does this relationship vary depending on whether the player won or lost the match?

```python

won_df = df[df['Result'] == 1]
lost_df = df[df['Result'] == 0]

plt.scatter(won_df['UFE.1'], won_df['Result'], label='Matches Won', alpha = 0.5)
plt.scatter(lost_df['UFE.1'], lost_df['Result'], label='Matches Lost', alpha = 0.6)


plt.title('Relationship between UFE.1 and Tournament Success')
plt.xlabel('Unforced Errors by player 1')
plt.ylabel('Tournament Result')
plt.legend()
plt.ylim(-1,3)

plt.show()
```

Question 2: Do players who serve more aces (ACE) tend to have higher first serve percentages (FSP) compared to players who serve fewer aces?

```python

more_aces = df1[df1['ACE.1'] > df1['ACE.2']]
less_aces = df1[df1['ACE.1'] < df1['ACE.2']]

more_aces_fsp = more_aces['FSP.1'].mean()
less_aces_fsp = less_aces['FSP.1'].mean()

plt.bar(['More Aces', 'Less Aces'], [more_aces_fsp, less_aces_fsp])
plt.title('Relationship between Aces and First Serve Percentage')
plt.xlabel('Aces')
plt.ylabel('Average First Serve Percentage')

plt.show()
```

Question 3: Is there a relationship between a player's first serve speed (FSP.1) and their success in the tournament, and does this relationship vary depending on the round of the tournament?

Question 4: Cluster dataset according to their serve data and how it is helpful?

```python

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


serve_df_french_w =df_french_w[['FSP.1', 'FSW.1']]

kmeans = KMeans(n_clusters=4, random_state= 0)
kmeans.fit(serve_df_french_w)

plt.scatter(serve_df_french_w['FSP.1'], serve_df_french_w['FSW.1'], c=kmeans.labels_, cmap='gist_rainbow', s =100, alpha = 0.7)
x = kmeans.cluster_centers_
plt.scatter(x[:,0], x[:,1], marker = "*", c = "r", s = 100, label = "Centroid")

plt.xlabel('First Serve Points Won Percentage')
plt.ylabel('First Serve Points Won')
plt.legend()
plt.show()
```

Question5: Is there a linear relationship between aces won by player 1 and the result? Design a model that can predict the possibility of winning the game by player 1 given the number of aces won?

What is the probability of wining a round by player 1 given that the number of aces won by player 1 is 10?

```python
from sklearn.linear_model import LinearRegression

cols = ['ACE.1', 'Result']

df_us_m = df_us_m[cols].dropna()

X = df_us_m[['ACE.1']]
y = df_us_m[['Result']]

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

plt.scatter(X, y, s=90, alpha = 0.5, c= "r")
plt.plot(X, y_pred, color='b', linewidth = 3, label = "Predicted line" )
plt.legend()
plt.ylim(-1, 2)
plt.xlabel('Aces won by player 1')
plt.ylabel('Result')
# plt.title("Relationship between aces won by player 1 and result")
plt.show()
m = (model.coef_)
c = (model.intercept_)

print(f"The slope and the intercept of the predicted line is :{m}, {c} resepctively.")

# What is the probability of wining a round by player 1 given that the number of 
# aces won by player 1 is 10?
p = model.predict(np.array(10).reshape(1, -1))
print("The probability of wining a round by player 1 given that the number of aces won by player 1 is 10: ",p[0][0])
```

Question 6: What is the relationship between a player's unforced errors (UFE.1) and their success rate in winning break points (BPW.1)?

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

df_us_w = df_us_w[['UFE.1', 'BPW.1']]

df_us_w = df_us_w.dropna()

sns.scatterplot(data=df_us_w, x='UFE.1', y='BPW.1',c = "r" ,s= 150, alpha = 0.7)
plt.title('Relationship between UFE.1 and BPW.1')
plt.xlabel('Unforced Errors')
plt.ylabel('Break Points Won %')

X = df_us_w[['UFE.1']]
y = df_us_w['BPW.1']
reg = LinearRegression().fit(X, y)

y_predicted = reg.predict(X)

plt.plot(X, y_predicted)
plt.show()
print('The coffecient of the linear regression model is:', reg.coef_[0])
print('the R-squared value is:', reg.score(X, y))
```

Question 7: What is the winning percentage (Result) of players who win more than 40% of their second serve points (SSP.1) and have less than 30 unforced errors (UFE.1) per match? 

```python
import seaborn as sns

df_wimbledon_m = df_wimbledon_m[['Result', 'SSP.1', 'UFE.1']]

df_filtered = df_wimbledon_m[(df_wimbledon_m['SSP.1'] > 40) & (df_wimbledon_m['UFE.1'] < 30)]
winning_percentage = df_filtered['Result'].mean() * 100

print('The winning percentage of players who win more than 40% of their second serve points and have less than 30 unforced errors is:', round(winning_percentage, 2), '%')

sns.histplot(data=df_filtered, x='SSP.1', kde=True)
plt.title('Distribution of SSP.1 values')
plt.xlabel('Second Serve Points Won %')
plt.show()
```

Question 8: What is the average number of winners (WNR.1) hit by players who won the match in FNL.1 = 2 and had a first serve percentage (FSP.1) of at least 60%? Visualize the distribution of these winner shots using a histogram.

```python
import pandas as pd
import matplotlib.pyplot as plt

filtered_data = df_wimbledon_w[(df_wimbledon_w['FNL.1'] == 2) & (df_wimbledon_w['FSP.1'] >= 60)]

avg_winners = filtered_data['WNR.1'].mean()
print("The average number of winners hit by players who won the match in straight sets and had a first serve percentage of at least 60%:", avg_winners)

sns.histplot(data = filtered_data, x = 'WNR.1', bins=10, color = "orange")
plt.title('Distribution of Winners')
plt.xlabel('Number of Winners')
plt.ylabel('Frequency')
plt.show()
```
