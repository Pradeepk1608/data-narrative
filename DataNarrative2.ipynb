# AAUP 


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df_aaup = pd.read_csv("http://lib.stat.cmu.edu/datasets/colleges/aaup.data")
df_aaup.columns =  columns = ['FICE', 'College name', 'State', 'Type','Average salary - full professors', 'Average salary - associate professors','Average salary - assistant professors', 'Average salary - all ranks','Average compensation - full professors', 'Average compensation - associate professors','Average compensation - assistant professors', 'Average compensation - all ranks',  'Number of full professors', 'Number of associate professors','Number of assistant professors', 'Number of instructors', 'Number of faculty - all ranks']
df_aaup.head()
```

Question1: Is there a relationship between the average compensation of full professors and the average compensation of all faculty ranks combined?


```python

plt.scatter(df_aaup["Average compensation - full professors"], df_aaup["Average compensation - all ranks"], color = "tab:green", s = 20)
plt.xlabel("Average compensation - full professors")
plt.ylabel("Average compensation - all ranks")

plt.show()

```

Question2: Does the average compensation of faculty members differ between institutions that offer primarily undergraduate degrees versus those that offer graduate and doctoral degrees?

```python
import matplotlib.pyplot as plt
import pandas as pd

mean_salary = df_aaup.groupby('Type')['Average salary - all ranks'].mean()
# print(mean_salary)
plt.bar(mean_salary.index, mean_salary.values)

plt.xlabel('Institution Types')
plt.ylabel('Mean Salary')
plt.title('Mean Salary Comparison between Research and Non-Research Institutions')
plt.show()
```

Question3: Which state has the maximum and least number of institutes and colleges?

```python
states_freq = df_aaup["State"].value_counts()
# print(states_freq)
states_freq.plot(kind = "bar", figsize= (15, 8))

plt.xlabel('States')
plt.ylabel('Number of institutes')
plt.title('Number of institutes corresponding to the each state')
plt.show()
```

Question4: Is there a difference in salaries and compensation between different states of the United States?

```python

# To calculate the mean salaries and compensations for each state
state_mean = df_aaup.groupby('State').mean()[['Average salary - all ranks', 'Average compensation - all ranks']]

state_mean.plot(kind='bar', figsize=(15, 8))

plt.xlabel('State')
plt.ylabel('Mean Salary/Compensation')
plt.title('Mean Salaries and Compensations by State')
plt.show()
```

Question5: Do institutions with a higher number of faculty members tend to have higher numbers of instructors as well?

```python
Num_faculty = df_aaup['Number of faculty - all ranks']
Num_instructors = df_aaup['Number of instructors']


plt.scatter(Num_faculty, Num_instructors, color = "m", s = 12)
plt.xlabel('Number of faculty members')
plt.ylabel('Number of instructors')
plt.title('Relationship between faculty members and instructors')
plt.show()
```

# USnews

```python

url = "http://lib.stat.cmu.edu/datasets/colleges/usnews.data"
df_us = pd.read_csv(url)
df_us.columns = ['FICE', 'College name', 'State', 'Public/private', 'Average Math SAT score', 'Average Verbal SAT score', 'Average Combined SAT score', 'Average ACT score', 'First quartile - Math SAT', 'Third quartile - Math SAT', 'First quartile - Verbal SAT','Third quartile - Verbal SAT','First quartile - ACT', 'Third quartile - +ACT','Number of applications received','Number of applicants accepted','Number of new students enrolled','Pct. new students from top 10% of H.S. class','Pct. new students from top 25% of H.S. class','Number of fulltime undergraduates',"Number of parttime undergraduates","In-state tuition","Out-of-state tuition","Room and board costs","Room costs", 'Board costs', 'Additional fees', 'Estimated book costs', 'Estimated personal spending', 'Pct. of faculty with Ph.D.s', 'Pct. of faculty with terminal degree', 'Student/faculty ratio', 'Pct.alumni who donate', 'Instructional expenditure per student', 'Graduation rate']

# df_us = df_us.dropna()
df_us.head()
```

Question6: Do private institutes have higher graduation rates? 

```python
df_us = df_us.dropna()

public_institute = df_us[df_us["Public/private"] == 1]
public_grad_rate = public_institute["Graduation rate"][:426].sort_values()

x = list(range(426))

private_institute = df_us[df_us["Public/private"] == 2] 
PRIVATE_grad_rate = private_institute["Graduation rate"][:426].sort_values()

# plt.grid()
plt.ylim(0, 150)
plt.plot(x, public_grad_rate)
plt.plot(x, PRIVATE_grad_rate)

plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)

plt.legend(["Private institute" , "Public institute"])
plt.ylabel("Graduation Rate")
plt.xlabel("Index")
plt.title("Graduation Rate of private and public institutions")
```

Question7: Do states which have higher graduation rate have higher Instructional expenditure per student?

```python
# df_us = df_us.dropna()
# plt.grid()

plt.scatter( df_us['Instructional expenditure per student'],df_us['Graduation rate'],color = "m",  s = 15)
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)

# plt.locator_params(axis='y', nbins=10) #number of ticks for y-axis
# plt.locator_params(axis='x', nbins=20) #number of ticks for x-axis
plt.xlim(0, 1000)
plt.xlabel('Instructional expenditure per student')
plt.ylabel('Graduation Rate')
plt.title('Relationship between Instructional expenditure per student and Graduation rate ')
plt.show()
```

Question 8: Which university has the maximum number of applications received?

```python
df_us= df_us[df_us["Number of applications received"]!= "*"]

df_us["Number of applications received"] = pd.to_numeric(df_us['Number of applications received'])


max_app_received_index = df_us['Number of applications received'].idxmax()
max_app_received_university = df_us.loc[max_app_received_index]['College name']
located_at = df_us.loc[max_app_received_index]['State']
print(f"The university with the maximum number of applications received is {max_app_received_university} which is located at {located_at}")

import matplotlib.pyplot as plt

# Sort the dataframe by number of applications received
df_sorted = df_us.sort_values(by='Number of applications received', ascending=False)

# Create a bar plot of the number of applications received by each college
plt.figure(figsize=(10,6))
plt.bar(df_sorted['College name'][:10], df_sorted['Number of applications received'][:10])
plt.xticks(rotation= 90)
plt.xlabel('College')
plt.ylabel('Number of applications received')
plt.title('Top 10 Colleges by Number of Applications Received')
plt.show()
```

Question 9: Do colleges with higher estimated book costs tend to have higher estimated personal spending?

```python
df_us= df_us[df_us["Estimated book costs"]!= "*"]
df_us["Estimated book costs"] = pd.to_numeric(df_us['Estimated book costs'])

df_us= df_us[df_us["Estimated personal spending"]!= "*"]
df_us["Estimated personal spending"] = pd.to_numeric(df_us['Estimated personal spending'])

x = df_us['Estimated book costs']
y = df_us['Estimated personal spending']
# plt.grid()

plt.scatter(x, y)
plt.xlabel('Estimated book costs')
plt.ylabel('Estimated personal spending')
plt.title('Relationship between Estimated Book Costs and Estimated Personal Spending')
plt.show()
```

Question 10: What is the probability that a randomly selected college in the US has a graduation rate above 80 percent?

```python
df_us= df_us[df_us["Graduation rate"]!= "*"]

df_us["Graduation rate"] = pd.to_numeric(df_us['Graduation rate'])

plt.xlim(0, 110)
plt.hist(df_us['Graduation rate'], bins=20)
plt.axvline(x=80, color='r', linestyle='--')
plt.xlabel('Graduation rate')
plt.ylabel('Frequency')
plt.show()
n_above_80 = len(df_us[df_us['Graduation rate'] > 80])
total = len(df_us)
prob_above_80 = n_above_80 / total

print(f"The probability of a randomly selected college in the US having a graduation rate above 80% is {prob_above_80}")
```
